```python
import networkx as nx
```

```python
my_graph = nx.Graph()
edges = nx.read_edgelist('/home/bella/Dropbox/Master-Thesis/MA_Thesis_Program/edges.txt')
nodes = nx.read_adjlist("/home/bella/Dropbox/Master-Thesis/MA_Thesis_Program/nodes.txt")
```

```python
my_graph.add_edges_from(edges.edges())
my_graph.add_nodes_from(nodes)
```

```python
nx.draw(my_graph, with_labels=True, font_size='10')
```
